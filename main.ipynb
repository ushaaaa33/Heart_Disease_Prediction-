```python
def plot_boxplot(df, column_name):
    
    plt.figure(figsize=(5,3))
    sns.boxplot(y=df[column_name])
    plt.title(f"Boxplot of {column_name}")
    plt.ylabel(column_name)
    plt.show()
```

```python
#choosing  x and y
x=df.drop('target',axis=1)
y=df.target
```

```python
x
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2,random_state=10)
```

```python
xtest
```

```python
ytest
```

```python
#model fitting
#creating logistric regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
#evaluation
# .predict()
yprediction=model.predict(xtest)
yprediction
```

```python
# to get accuracy
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score,f1_score
accuracy=accuracy_score(ytest,yprediction)
print(f'Accuracy={accuracy:.2%}')
```

```python
# creating confusion metrix
cm=confusion_matrix(ytest,yprediction)
cm
```

```python
precission=precision_score(ytest,yprediction)
print(f'Pricission={precission:.2%}')
```

```python
recal=recall_score(ytest,yprediction)
print(f'Recal_score={recal:.2%}')
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,yprediction))
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
```

```python
# model fitting using random forest
ranfor_model = RandomForestClassifier()
ranfor_model.fit(xtrain, ytrain)
```

```python
# accuracy
ranfor_model_score = ranfor_model.score(xtest, ytest)
ranfor_model_score
```

```python
# model fitting using decision tree
dectree_model = DecisionTreeClassifier(max_depth=5)
dectree_model.fit(xtrain, ytrain)
```

```python
dectree_model_score = dectree_model.score(xtest, ytest)
dectree_model_score
```

```python
gbc_model = GradientBoostingClassifier()
gbc_model.fit(xtrain, ytrain)
```

```python
gbc_model_score = gbc_model.score(xtest, ytest)
gbc_model_score
```

```python
gaussnb_model = GaussianNB()
gaussnb_model.fit(xtrain, ytrain)
```

```python
gaussnb_model_score = gaussnb_model.score(xtest, ytest)
gaussnb_model_score
```

```python
knn_model = KNeighborsClassifier(n_neighbors=9)
knn_model.fit(xtrain, ytrain)
```

```python
knn_model_score = knn_model.score(xtest, ytest)
knn_model_score
```

```python
mlp_model = MLPClassifier()
mlp_model.fit(xtrain, ytrain)
```

```python
mlp_model_score = mlp_model.score(xtest, ytest)
mlp_model_score
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
sr=StandardScaler()
xtrain_s=sr.fit_transform(xtrain)
```

```python
#model fitting
from sklearn.svm import SVC
svm_model=SVC(kernel='rbf')
svm_model.fit(xtrain_s, ytrain)
```

```python
svm_model_score = svm_model.score(xtest, ytest)
svm_model_score
```
